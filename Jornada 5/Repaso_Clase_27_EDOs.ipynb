# Repaso EDO's
* Dos cuerpos (simplificado)
> 1. Un cuerpo masivo '(M)' se ubica en el orígen.
> 2. Un cuerpo de menor masa '(m)' experimenta el campo gravitatorio de 'M'
Ley de gravitación universal
$$ \vec{F_g} = -G \frac{mM}{r^2} \hat{r} = -G \frac{mM}{r^3} \vec{r} $$
$$ \text{1. } \hat{r} = \frac{\vec{r}}{r}$$ 
$$ \text{2.} \textbf{ 2da ley newton } \vec{F_g} = m\cdot\vec{a} $$

$$ -G \frac{mM}{r^3}\vec{r} = m\cdot\dfrac{d\vec{v}}{dt}$$
Ec. de segundo orden.
$$ \dfrac{d\vec{v}}{dt} = -\frac{GM}{r^3}\vec{r} \text{          ;     }  \dfrac{\vec{r}}{dt} = \vec{v}

Sistema de EDO's de primer orden en forma Standard.
Vamos a resolver este sist. usando ***Euler***
$$ \dfrac{d\vec{v}}{dt} \simeq \frac{\vec{V}_{n+1} - \vec{V}_{n}}{\Delta t}$$
$$ \dfrac{d\vec{r}}{dt} \simeq \frac{\vec{r}_{n+1} - \vec{r}_{n}}{\Delta t}$$

$$\frac{\vec{V}_{n+1} - \vec{V}_{n}}{\Delta t} = - \frac{GM}{{r^3}_{n}}\vec{r}_{n}$$
$$\frac{\vec{r}_{n+1} - \vec{r}_{n}}{\Delta t} = \vec{V}_{n}$$
> Con esto tenemos nuestras escuaciones iterativas:
$$\vec{V}_{n+1} = \vec{V}_{n} + \Delta t \cdot -\frac{GM}{{r^3}_{n}} \vec{r}_{n}$$
$$\vec{r}_{n+1} = \vec{r}_{n} + \Delta{t}\cdot{\vec{V}_{n}}$$
$$ t_{n+1} = t_n + \Delta{t}

>Por ser de segundo orden, necesitamos 2 cond. iniciales.
$$\vec{r_0}=(x_0,0),\quad \vec{V_o} = (0,v_0),\quad t_0$$
>n=0)
$$\vec{V_1} = \vec{V_0} + \Delta{t}\cdot{\frac{GM}{{r^3}_{0}}} \vec{r_0}\quad,\quad \vec{r_1} = \vec{r_0} + \Delta{t}\cdot{\vec{V_0}}\quad,\quad t_1$$
>n=1)
$$\vec{V_2} = \vec{V_1} + \Delta{t}\cdot{\frac{GM}{{r^3}_{1}}} \vec{r_1}\quad,\quad \vec{r_2} = \vec{r_1} + \Delta{t}\cdot{\vec{V_1}}\quad,\quad t_2$$
